```python
import requests
import datetime
from datetime import datetime
from itertools import groupby
from operator import itemgetter

class Meetings():
    base_url = 'https://ct-mock-tech-assessment.herokuapp.com/'
    calls = []
    
    def __init__(self): 
        self.data = None
        self.calls.append(self)
        
    def __repr__(self):
        return f'<APICall: works>'
    def run(self):
        response = requests.get(f'{self.base_url}')
        self.data = response.json()
    def country_dict(count, emails, country, date):
        invitation = {
            "AttendeeCount":count,
            "Attendees": emails,
            "Name": country,
            "Start Date": date
        }
        return invitation
    def most_common(lst):
        return max(set(lst), key=lst.count)
    def ranges(nums):
        #nums = sorted(s(nums))
        gaps = [[s, e] for s, e in zip(nums, nums[1:]) if s+1 < e]
        edges = iter(nums[:1] + sum(gaps, []) + nums[-1:])
        return list(zip(edges, edges))
    def convert(country):
        dates = []
        good_dates = []
        for citizens in country:
            for date in citizens["availableDates"]:
                #print(date)
                dates.append(date)
                consecutive_dates = [datetime.strptime(d, "%Y-%m-%d") for d in dates]
                date_ints = ([d.toordinal() for d in consecutive_dates])
        return date_ints
    def convert_back(best_dates):
        meeting_dates = []
        for days in best_dates:
            meeting_dates.append(datetime.fromordinal(days))
        #print(meeting_dates)
        str_dates = []  

        #print(meeting_dates)
        for dayss in meeting_dates:
            str_dates.append(dayss.strftime("%Y-%m-%d"))
            #print(dayss.strftime("%Y-%m-%d"))
        return str_dates
#print(str_dates)
    def emails(country, str_dates):
        emails = []
        for email in country:
            if str_dates[0] and  str_dates[1] in email['availableDates']:
       
                emails.append(email['email'])
        return emails
#freedom_units = len(american_emails)     
    #countries
united_states  = []
ireland = []
spain = []
mexico = []
canada = []
singapore = []
japan = []
united_kingdom = []
france = []

invitation = []

base_url = 'https://ct-mock-tech-assessment.herokuapp.com/'
    
meetings = Meetings()
meetings.run()
meetings.calls
partners = meetings.data

partner = partners['partners']

for peeps in partner: 
    #country = partners['partners'][0]['country']
    if peeps['country'] == 'United States':
        united_states.append(peeps)
    elif peeps['country'] == "France":
        france.append(peeps)
    elif peeps['country'] == 'Ireland':
        ireland.append(peeps)
    elif peeps['country'] == 'Spain':
        spain.append(peeps)
    elif peeps['country'] == 'Mexico':
        mexico.append(peeps)
    elif peeps['country'] == "Canada":
        canada.append(peeps)
    elif peeps['country'] == 'Singapore':
        singapore.append(peeps)
    elif peeps['country'] == 'Japan':
        japan.append(peeps)
    elif peeps['country'] =='United Kingdom':
        united_kingdom.append(peeps)
# united states

american_date_ints = Meetings.convert(united_states)
american_common_dates = Meetings.ranges(american_date_ints)
american_best_dates = Meetings.most_common(american_common_dates)
american_str_dates = Meetings.convert_back(american_best_dates)
american_invitation = Meetings.country_dict(len(Meetings.emails(united_states, american_str_dates)), Meetings.emails(united_states, american_str_dates), "United States", american_str_dates[0])
#american_invitation

#France
french_date_ints = Meetings.convert(france)
french_common_dates = Meetings.ranges(french_date_ints)
french_best_dates = Meetings.most_common(french_common_dates)
french_str_dates = Meetings.convert_back(french_best_dates)
french_invitation = Meetings.country_dict(len(Meetings.emails(france, french_str_dates)), Meetings.emails(france, french_str_dates), "France", french_str_dates[0])
#french_invitation

#Ireland
irish_date_ints = Meetings.convert(ireland)
irish_common_dates = Meetings.ranges(irish_date_ints)
irish_best_dates = Meetings.most_common(irish_common_dates)
irish_str_dates = Meetings.convert_back(irish_best_dates)
irish_invitation = Meetings.country_dict(len(Meetings.emails(ireland, irish_str_dates)), Meetings.emails(ireland, irish_str_dates), "Ireland", irish_str_dates[0])
#irish_invitation

#Spain
spanish_date_ints = Meetings.convert(spain)
spanish_common_dates = Meetings.ranges(spanish_date_ints)
spanish_best_dates = Meetings.most_common(spanish_common_dates)
spanish_str_dates = Meetings.convert_back(spanish_best_dates)
spanish_invitation = Meetings.country_dict(len(Meetings.emails(spain, spanish_str_dates)), Meetings.emails(spain, spanish_str_dates), "Spain", spanish_str_dates[0])
#spanish_invitation

#Mexico
mexican_date_ints = Meetings.convert(mexico)
mexican_common_dates = Meetings.ranges(mexican_date_ints)
mexican_best_dates = Meetings.most_common(mexican_common_dates)
mexican_str_dates = Meetings.convert_back(mexican_best_dates)
mexican_invitation = Meetings.country_dict(len(Meetings.emails(mexico, mexican_str_dates)), Meetings.emails(mexico, mexican_str_dates), "Mexico", mexican_str_dates[0])
#mexican_invitation

#Canada
canadian_date_ints = Meetings.convert(canada)
canadian_common_dates = Meetings.ranges(canadian_date_ints)
canadian_best_dates = Meetings.most_common(canadian_common_dates)
canadian_str_dates = Meetings.convert_back(canadian_best_dates)
canadian_invitation = Meetings.country_dict(len(Meetings.emails(canada, canadian_str_dates)), Meetings.emails(canada, canadian_str_dates), "Canada", canadian_str_dates[0])
#canadian_invitation

#Singapore
singapore_date_ints = Meetings.convert(singapore)
singapore_common_dates = Meetings.ranges(singapore_date_ints)
singapore_best_dates = Meetings.most_common(singapore_common_dates)
singapore_str_dates = Meetings.convert_back(singapore_best_dates)
singapore_invitation = Meetings.country_dict(len(Meetings.emails(singapore, singapore_str_dates)), Meetings.emails(singapore, singapore_str_dates), "Singapore", singapore_str_dates[0])
#singapore_invitation

#Japan
japanese_date_ints = Meetings.convert(japan)
japanese_common_dates = Meetings.ranges(japanese_date_ints)
japanese_best_dates = Meetings.most_common(japanese_common_dates)
japanese_str_dates = Meetings.convert_back(japanese_best_dates)
japanese_invitation = Meetings.country_dict(len(Meetings.emails(japan, japanese_str_dates)), Meetings.emails(japan, japanese_str_dates), "Japan", japanese_str_dates[0])
#japanese_invitation

#United Kingdom
kingdom_date_ints = Meetings.convert(united_kingdom)
kingdom_common_dates = Meetings.ranges(kingdom_date_ints)
kingdom_best_dates = Meetings.most_common(kingdom_common_dates)
kingdom_str_dates = Meetings.convert_back(kingdom_best_dates)
kingdom_invitation = Meetings.country_dict(len(Meetings.emails(united_kingdom, kingdom_str_dates)), Meetings.emails(united_kingdom, kingdom_str_dates), "United Kingdom", kingdom_str_dates[0])
#kingdom_invitation

invitation ={'data': [american_invitation, irish_invitation, french_invitation, spanish_invitation, mexican_invitation, canadian_invitation, singapore_invitation, japanese_invitation, kingdom_invitation]}
#invitation
new_data = {'data' : invitation}
print(new_data)
x = requests.post(base_url,data=invitation)
print(x.status_code)
print(x.text)

```
